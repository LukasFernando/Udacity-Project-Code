## Question 1: How much is the average salary for top five country?

Here we'll import some libraries

```python
import pandas as pd              # Use to read the data 
import matplotlib.pyplot as plt  # To create a graphic 
```

Reading the file 

```python
df_public = pd.read_csv('survey_results_public.csv') # Reading the data 
df_public.head()  # Showing some (five) lines 
```

## Clean data 

We'll take just two columns: Country and Salary 

```python
df_public_filtered = df_public[['Country', 'Salary']]  # Selecting the two column 
df_public_filtered = df_public_filtered.dropna(axis=0, how='any')  # Dropping all the lines that have an empty value
df_public_filtered.head()
```

## Modeling the data

Grouping by using the column "Country" and getting the average for each country 

```python
df_public_group = df_public_filtered.groupby(df_public_filtered['Country']).mean()  # Grouping by "Country" and getting the average
df_public_group.head()
```

Sorting the Salary to get the five biggest

```python
# Here the high salary will be on top, so we just use the "head()" to get the five highest salary
df_filt = df_public_group.sort_values(by='Salary', ascending=False).head()
df_filt
```

## Graphic

The graphic used is bar to show the five country that have the highest salary

```python
df_filt.plot(kind='bar')
```

Now we can see that Bermuda is the country that has the highest average salary. So if you start a new office but don't have a lot of money, you can start a new office in some country where the salary is not so expensive for your company.
