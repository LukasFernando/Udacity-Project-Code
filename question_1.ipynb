Question 1: How much is the average salary for top five country?

```python
# imports 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading the file 
df_public = pd.read_csv('survey_results_public.csv')
```

```python
df_public.head()
```

```python
# Filtering to get just the Country and Salary.
df_public_filtered = df_public[['Country', 'Salary']]
df_public_filtered = df_public_filtered.dropna(axis=0, how='any')
df_public_filtered.head()
```

```python
# Grouping by Country and getting the average
df_public_group = df_public_filtered.groupby(df_public_filtered['Country']).mean()
df_public_group.head()
```

```python
# Sorting the Salary to get the five biggest
df_filt = df_public_group.sort_values(by='Salary', ascending=False).head()
df_filt
```

```python
df_filt.plot(kind='bar')
```
